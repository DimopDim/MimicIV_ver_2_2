```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File paths
xgboost_file_path = 'csv/testing/xgboost_performance.csv'
naive_file_path = 'csv/testing/naive_bayes_performance.csv'
decision_tree_file_path = 'csv/testing/decision_tree_performance.csv'

# Read CSV files
xgboost_df = pd.read_csv(xgboost_file_path, index_col=0)
naive_df = pd.read_csv(naive_file_path, index_col=0)
decision_tree_df = pd.read_csv(decision_tree_file_path, index_col=0)

# Extracting relevant values
algorithms = ['XGBoost', 'Naive Bayes', 'Decision Tree']
metrics = ['accuracy', 'absolute_error', 'root_mean_squared_error', 'squared_error', 'correlation']

# Prepare data for plotting
data = np.array([
    [xgboost_df.loc[metric][0] for metric in metrics],
    [naive_df.loc[metric][0] for metric in metrics],
    [decision_tree_df.loc[metric][0] for metric in metrics]
])

# Plotting
bar_width = 0.2
index = np.arange(len(metrics))

fig, ax = plt.subplots(figsize=(10, 6))

for i, algorithm in enumerate(algorithms):
    ax.bar(index + i * bar_width, data[i], bar_width, label=algorithm)

ax.set_xlabel('Metrics')
ax.set_ylabel('Performance Values')
ax.set_title('Algorithm Performance Comparison')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(metrics)
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File paths
xgboost_file_path = 'csv/testing/xgboost_performance.csv'
naive_file_path = 'csv/testing/naive_bayes_performance.csv'
decision_tree_file_path = 'csv/testing/decision_tree_performance.csv'

# Read CSV files
xgboost_df = pd.read_csv(xgboost_file_path, index_col=0)
naive_df = pd.read_csv(naive_file_path, index_col=0)
decision_tree_df = pd.read_csv(decision_tree_file_path, index_col=0)

# Extracting relevant values
algorithms = ['XGBoost', 'Naive Bayes', 'Decision Tree']
metrics = ['accuracy', 'absolute_error', 'root_mean_squared_error', 'squared_error', 'correlation']

# Prepare data for plotting
data = np.array([
    [xgboost_df.loc[metric][0] for metric in metrics],
    [naive_df.loc[metric][0] for metric in metrics],
    [decision_tree_df.loc[metric][0] for metric in metrics]
])

# Create a radar chart for each algorithm
for i, algorithm in enumerate(algorithms):
    values = data[i].tolist()

    # Adjust the number of angles to match the number of metrics
    angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False).tolist()

    # Plot
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    ax.plot(angles, values, color='blue', linewidth=2, linestyle='solid', marker='o')
    ax.fill(angles, values, color='blue', alpha=0.4)

    # Set labels and title
    ax.set_yticklabels([])
    ax.set_xticks(angles)
    ax.set_xticklabels(metrics)
    ax.set_title(f'Radar Chart - {algorithm}')

    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File paths
xgboost_file_path = 'csv/testing/xgboost_performance.csv'
naive_file_path = 'csv/testing/naive_bayes_performance.csv'
decision_tree_file_path = 'csv/testing/decision_tree_performance.csv'

# Read CSV files
xgboost_df = pd.read_csv(xgboost_file_path, index_col=0)
naive_df = pd.read_csv(naive_file_path, index_col=0)
decision_tree_df = pd.read_csv(decision_tree_file_path, index_col=0)

# Extracting relevant values
algorithms = ['XGBoost', 'Naive Bayes', 'Decision Tree']
metrics = ['accuracy', 'absolute_error', 'root_mean_squared_error', 'squared_error', 'correlation']

# Prepare data for plotting
data = np.array([
    [xgboost_df.loc[metric][0] for metric in metrics],
    [naive_df.loc[metric][0] for metric in metrics],
    [decision_tree_df.loc[metric][0] for metric in metrics]
])

# Create a radar chart for all algorithms
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Adjust the number of angles to match the number of metrics
angles = np.linspace(0, 2 * np.pi, len(metrics), endpoint=False).tolist()

for i, algorithm in enumerate(algorithms):
    values = data[i].tolist()

    # Plot with different colors for each algorithm
    color = plt.cm.viridis(i / len(algorithms))  # Use a colormap for different colors
    ax.plot(angles, values, linewidth=2, linestyle='solid', marker='o', label=algorithm, color=color)
    ax.fill(angles, values, alpha=0.4, color=color)

# Set labels and title
ax.set_yticklabels([])
ax.set_xticks(angles)
ax.set_xticklabels(metrics)
ax.set_title('Radar Chart - Algorithm Comparison')
ax.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# File paths
xgboost_file_path = 'csv/testing/xgboost_performance.csv'
naive_file_path = 'csv/testing/naive_bayes_performance.csv'
decision_tree_file_path = 'csv/testing/decision_tree_performance.csv'

# Read CSV files
xgboost_df = pd.read_csv(xgboost_file_path, index_col=0)
naive_df = pd.read_csv(naive_file_path, index_col=0)
decision_tree_df = pd.read_csv(decision_tree_file_path, index_col=0)

# Extracting relevant values
algorithms = ['XGBoost', 'Naive Bayes', 'Decision Tree']
metrics = ['accuracy', 'absolute_error', 'root_mean_squared_error', 'squared_error', 'correlation']

# Prepare data for plotting
data = np.array([
    [xgboost_df.loc[metric][0] for metric in metrics],
    [naive_df.loc[metric][0] for metric in metrics],
    [decision_tree_df.loc[metric][0] for metric in metrics]
])

# Create a DataFrame for heatmap plot
heatmap_df = pd.DataFrame(data.T, columns=algorithms, index=metrics)

# Melt the DataFrame for easier plotting
melted_df = heatmap_df.reset_index().melt(id_vars='index', var_name='Algorithm', value_name='Value')

# Create a heatmap using pivot_table
plt.figure(figsize=(10, 6))
heatmap_data = melted_df.pivot_table(index='index', columns='Algorithm', values='Value', aggfunc='mean')
sns.heatmap(heatmap_data, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
plt.title('Algorithm Performance Comparison - Heatmap')
plt.show()
```
