```python
import gc
import gzip
import pandas as pd

# Read the stroke_diagnosis.csv file
mean_df = pd.read_csv("CSV\Exports\datasets\whole_set\o01_final_mean_table.csv")
median_df = pd.read_csv("CSV\Exports\datasets\whole_set\o02_final_median_table.csv")
min_df = pd.read_csv("CSV\Exports\datasets\whole_set\o03_final_min_table.csv")
max_df = pd.read_csv("CSV\Exports\datasets\whole_set\o04_final_max_table.csv")
```

```python
# Merge the datasets based on the common columns
merged_df = pd.concat([mean_df, median_df, min_df, max_df], axis=0, ignore_index=True)

# Now, merged_df contains all the rows from the four datasets with common columns.
```

```python
display(merged_df)
```

```python
# Extract the 'hospital_expired_flag' column
hospital_expired_flag_column = merged_df.pop('hospital_expire_flag')

# Add it back as the last column
merged_df['hospital_expire_flag'] = hospital_expired_flag_column
```

```python
merged_df.to_csv('CSV\Exports\oThrombo_jadbio_.csv', index=False)
```

```python
filtered_df = merged_df[merged_df['Time_Zone'] == 8]

filtered_df = filtered_df.copy()
filtered_df.reset_index(drop=True, inplace=True)
filtered_df.drop(columns=['row_count', 'subject_id', 'hadm_id', 'Time_Zone'], inplace=True)
```

```python
display(filtered_df)
filtered_df.to_csv('CSV\Exports\oThrombo_24_hours.csv', index=False)
```

```python
# Calculate the index for splitting the DataFrame
split_index = int(0.7 * len(filtered_df))

# Split the DataFrame into training and test sets
train_df = filtered_df[:split_index]
test_df = filtered_df[split_index:]
```

```python
train_df.to_csv('CSV\Exports\oThrombo_24_hours_training_.csv', index=False)

test_df.to_csv('CSV\Exports\oThrombo_24_hours_test_.csv', index=False)
```

```python
# Columns to exclude from the count
columns_to_exclude = ['gender', 'age', 'language', 'marital_status', 'race']

# Calculate non-NaN counts for all columns except the excluded ones
non_nan_counts = train_df.loc[:, ~train_df.columns.isin(columns_to_exclude)].count()

# If you want to count the total number of non-NaN values in the entire DataFrame:
total_non_nan_count = non_nan_counts.sum()

print("Non-NaN counts in each column (excluding specified columns):")
print(non_nan_counts)
print("\nTotal count of non-NaN values in the entire DataFrame:")
print(total_non_nan_count)

```
