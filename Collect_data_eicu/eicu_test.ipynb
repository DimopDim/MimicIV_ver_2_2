```python
"""
Checking if the files have the same
values in the columns
"""

import pandas as pd

def are_csv_files_same_values(file_paths):
    # Read each CSV file into a DataFrame, starting from the second row
    dataframes = [pd.read_csv(file, header=None, skiprows=1) for file in file_paths]

    # Check if columns and values are the same
    for col in dataframes[0].columns:
        for i in range(1, len(dataframes)):
            if not dataframes[0][col].equals(dataframes[i][col]):
                return False

    return True

# List of CSV file paths
csv_files = ['CSV\Exports\datasets\Temp\o01_mean_table.csv',
             'CSV\Exports\datasets\Temp\o02_median_table.csv',
             'CSV\Exports\datasets\Temp\o03_min_table.csv',
             'CSV\Exports\datasets\Temp\o04_max_table.csv']

# Check if CSV files have the same values in columns (excluding headers)
if are_csv_files_same_values(csv_files):
    print("The CSV files have the same values in columns (excluding headers).")
else:
    print("The CSV files do not have the same values in columns (excluding headers).")
```
